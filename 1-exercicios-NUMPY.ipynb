```
import numpy as np
np.random.seed(10)
```

### 1. Criar um array com 10 números inteiros aleatórios entre 1 e 50 e calcular a média e o desvio padrão

```
array = np.random.randint(50, size=(10))

media = array.mean()
desvio = array.std()
media, desvio
```

### 2. Criar um array de 10 números aleatórios e calcular a raiz quadrada de todos os valores menores ou iguais a 0,5

```
array = np.random.rand(10)

array[array <= 0.5] = np.sqrt(array[array <= 0.5])

array
```

### 3. Criar uma matriz 3D de tamanho 3x3x3 com valores aleatórios e exibir o valor máximo de cada linha de cada matriz 2D dentro dessa matriz 3D

```
matriz = np.random.rand(3, 3, 3)

max_value = np.amax(matriz, axis=1)

print(
f"""
Matrize:
{matriz}

Matriz máxima:
{max_value}
""")
```

### 4. Criar uma matriz de tamanho 5x5 com valores aleatórios e subtrair a média de cada linha dessa matriz

```
matriz = np.random.rand(5, 5)
print(f"Matrizes: {matriz}")
matriz_media = np.mean(matriz, 1)

subtracao = matriz - matriz_media[:, np.newaxis]
subtracao
```

### 5. Criar uma matriz de tamanho 4x4 com valores aleatórios e calcular a média e o desvio padrão das diagonais da matriz

```
matriz = np.random.rand(4, 4)

diag_1 = np.diag(matriz)
diag_2 = np.diag(np.fliplr(matriz))

media_d1 = np.mean(diag_1)
media_d2 = np.mean(diag_2)

desv_d1 = np.std(diag_1)
desv_d2 = np.std(diag_2)

print(
f"""
Matriz principal:
{matriz}
Diagonal 1:
{diag_1}
Média:
{media_d1:.2f}
Desvio:
{desv_d1:.2f}

Diagonal 2:
{diag_2}
Média:
{media_d2:.2f}
Desvio:
{desv_d2:.2f}
"""
)
```
